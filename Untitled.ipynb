```python
def distancia_coseno(u,v):
    return (1-np.cos(np.dot(u,v)))
```

```python
def getManhattan(xList,yList):
    res = 0
    for i in range(len(xList)+1):
        res = res + (xList[0]-yList[0])
    return res
```

```python
def Euclediana(a,b):
    return np.linalg.norm(a-b)
```
